#### 1. Problem Example 
- 두 수를 뽑은다음에 더해서 리스트를 오름차순으로 리턴
- 1분 미만 소요


```python

def solution(numbers):
    
    combination = list(permutations(numbers, 2))
    answer = [x[0]+x[1] for x in combination]
    
    answer = list(set(answer))
    answer = sorted(answer)

    return answer
```
