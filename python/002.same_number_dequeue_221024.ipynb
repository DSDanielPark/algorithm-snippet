#### 1 problem
- [1,1,3,3,0,1,1] to [1,3,0,1]
- [4,4,4,3,3] to [4,3]



1.1 Solution


```python

def solution(arr):
    answer = []
    temp = None
    for k in arr:
        if k != temp: 
            answer.append(k)
        temp = k
        
    return answer
```

1.2 The other Solution

```python
def no_continuous(s):
    a = []
    for i in s:
        if a[-1:] == [i]: continue
        a.append(i)
    return a

# 아래는 테스트로 출력해 보기 위한 코드입니다.
print( no_continuous( "133303" ))
```
