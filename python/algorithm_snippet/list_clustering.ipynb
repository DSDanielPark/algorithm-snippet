```python
points = [0.1, 0.31,  0.32, 0.45, 0.35, 0.40, 0.5 ]
```

```python
def get_clusted_list(target_list: list, eps: float) -> list:
    '''
    특정 간격(eps)을 기준으로 해당 리스트를 클러스터링 함

    target_list: 클러스터링 할 1d list
    eps: 클러스터링할 때 구별할 간격
    
    return: 2d list
    '''

    clusters = []
    points_sorted = sorted(target_list)
    curr_point = points_sorted[0]
    curr_cluster = [curr_point]
    for point in points_sorted[1:]:
        if point <= curr_point + eps:
            curr_cluster.append(point)
        else:
            clusters.append(curr_cluster)
            curr_cluster = [point]
        curr_point = point
    clusters.append(curr_cluster)

    return clusters
```

```python
get_clusted_list(points, 0.2)
```
