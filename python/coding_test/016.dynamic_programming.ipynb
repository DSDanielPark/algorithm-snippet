#### 1 Problem `1`
- 2개씩 중복되면 제거 (3개는 안됨) <br>
예를 들어, 문자열 S = baabaa 라면 <br>
b aa baa → bb aa → aa → <br>


[example]
- baabaa >> 1
- cdcd >> 0
 


1.1 My Solution

solution a

```python
def solution(s):
    def del_dupli(t1):
        return [t1[i] for i in range(len(t1)) if t1[i:i+1] != t1[i+1:i+2] and t1[i-1:i] != t1[i:i+1] ]

    while True:
        nex = del_dupli(s)
        if nex == s:
            break
        s = nex
        
    if len(nex) == 0:
        return 1
    return 0
```
