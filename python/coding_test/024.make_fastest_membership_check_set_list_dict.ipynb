```python
#iteration 비교
```

```python
#using list
def method1(A):
    list1 = A[0]
    list2 = A[1]
    
    not_common_list = [x for x in list1 if x not in list2] + [x for x in list2 if x not in list1]
    
    return list1[-1] in not_common_list
```

```python
#using set
def method2(A):
    set1 = set(A[0])
    set2 = set(A[1])
    
    not_common_set = list(set1-set2) + list(set2-set1)
    
    return list(set1)[-1] in not_common_set
```

```python
import functools
import itertools
import operator
import numpy as np
import perfplot
import random
     
b = perfplot.bench(
    setup= lambda n: [[random.randint(1,1000000) for _ in range(1000)], [random.randint(1,1000000) for _ in range(500)]],
    # setup=lambda n: [list(range(n))] * 10,
    kernels=[
        method1,
        method2,
    ],
    n_range=[2 ** k for k in range(16)],
    xlabel="num lists (of length 1000)",
    # xlabel="len lists (10 lists total)"
)
b.save("out.png")
b.show()
```
