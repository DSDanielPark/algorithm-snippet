#### 1 Problem `1`
- LRU(Least Recently Used) 페이지 교체 알고리즘, 도시와 캐시사이즈가 인풋으로 주어짐
- hit = +1
- miss = +5
[example]

1.1 My Solution (Success)


```python
def solution(cacheSize, cities):
    answer = 0
    cache = []
    
    cities = [c.lower() for c in cities]
    
    if cacheSize == 0:
        return len(cities) * 5
    else:
        for c in cities:
            if not c in cache:
                if len(cache) < cacheSize:
                    cache.append(c)
                    answer += 5
                else:
                    cache.pop(0)
                    cache.append(c)
                    answer += 5
            else:
                cache.pop(cache.index(c))
                cache.append(c)
                answer += 1
                
    return answer
                
```
