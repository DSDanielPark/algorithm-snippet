```python
# you can write to stdout for debugging purposes, e.g.
# print("this is a debug message")
from collections import Counter

def solution1(A):
    temp_dict = Counter(A)
    unique_element = [k for k,v in temp_dict.items() if v==1]
    
    loc_list = []
    
    if len(unique_element)==0:
        return -1
    
    else:
        for u in unique_element:
            loc_list.append(A.index(u))
        return min(loc_list)
    
```

```python
solution(A)
```

```python
# you can write to stdout for debugging purposes, e.g.
# print("this is a debug message")

def solution2(A):
    elements = list(set(A))
    A_copy = A.copy()
    
    for e in elements:
        A_copy.pop(A_copy.index(e))
    unique_val = [A.index(e) for e in elements if e not in A_copy]
    
    if len(unique_val)==0:
        return -1
    else:
        return min(unique_val)
        
```

```python
import functools
import itertools
import operator

import numpy as np
import perfplot
```

```python
def forfor(a):
    return [item for sublist in a for item in sublist]


def sum_brackets(a):
    return sum(a, [])


def functools_reduce(a):
    return functools.reduce(operator.concat, a)


def functools_reduce_iconcat(a):
    return functools.reduce(operator.iconcat, a, [])


def itertools_chain(a):
    return list(itertools.chain.from_iterable(a))


def numpy_flat(a):
    return list(np.array(a).flat)


def numpy_concatenate(a):
    return list(np.concatenate(a))


def extend(a):
    out = []
    for sublist in a:
        out.extend(sublist)
    return out
```

```python
b = perfplot.bench(
    setup=lambda n: [list(range(10))] * n,
    # setup=lambda n: [list(range(n))] * 10,
    kernels=[
        forfor,
        sum_brackets,
        functools_reduce,
        functools_reduce_iconcat,
        itertools_chain,
        numpy_flat,
        numpy_concatenate,
        extend,
    ],
    n_range=[2 ** k for k in range(16)],
    xlabel="num lists (of length 10)",
    # xlabel="len lists (10 lists total)"
)
b.save("out.png")
b.show()
```
