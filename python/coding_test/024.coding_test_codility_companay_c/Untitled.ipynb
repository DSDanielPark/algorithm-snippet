```python
A = [1, 3, 6, 4, 1, 2]
B = [1, 2, 3]
C = [-1,-3]
```

```python
full_list = [x for x in range(1, len(A)+1)]
```

```python
full_list
```

```python
min([x for x in full_list if x not in A])
```

```python
def solution(A):
    plus_list = [x for x in A if x>0]
    if len(plus_list)==0:
        return 1
    else:
        full_list = [x for x in range(1, len(A)+1)]
        empty_list = [x for x in full_list if x not in A]
        if len(empty_list)==0:
            return max(A)+1
        else:
            return min(empty_list)
    
```

```python
print(solution(A))
print(solution(B))
print(solution(C))
```
