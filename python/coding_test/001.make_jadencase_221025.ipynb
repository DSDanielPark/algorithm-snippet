1.1 My Solution

```python
def solution(s):
    
    split_list = s.split(' ')
    split_list = [s for s in split_list]
    
    new_list = []
    for i, s in enumerate(split_list):
        if (len(s)!=0)and(s[0] != int):
            s = list(s)
            s = [k.lower() for k in s]
            s[0] = (s[0].upper())
            s=''.join(s)
            new_list.append(s)
        
        else:
            new_list.append(s)
        
        
    print(new_list)
    
    answer = ' '.join(new_list)
    
    return answer
```
