```python
def solution(lottos, win_nums):
    matched_list = [x for x in lottos if x in win_nums]
    
    #최저등수
    if len(matched_list) == 6:
        min_p = 1
    elif len(matched_list) == 5:
        min_p = 2
    elif len(matched_list) == 4:
        min_p = 3
    elif len(matched_list) == 3:
        min_p = 4
    elif len(matched_list) == 2:
        min_p = 5
    else:
        min_p = 6

    answer = []
    
    zero_list =  [x for x in lottos if x==0]
    not_matched_list = [x for x in win_nums if x not in lottos]
    
    if len(zero_list) <= len(not_matched_list):
        cnt = len(matched_list)+len(zero_list)
    else: 
        cnt = len(matched_list)+len(not_matched_list)
    
    if cnt == 6:
        max_p = 1
    elif cnt == 5:
        max_p = 2
    elif cnt == 4:
        max_p = 3
    elif cnt == 3:
        max_p = 4
    elif cnt == 2:
        max_p = 5
    else:
        max_p = 6
        
    answer.append(max_p)
    answer.append(min_p)
    return answer
```
