# 1. general using of class1 (1)

```python
class Test:
    def __init__(self):
        self.a = 3
        self.b = 6
        print('init')
        print(f'self instance is memorized: a={self.a} b={self.b}')
    
    def __call__(self):
        print('call')
        return Test.function(self, self.a, self.b)
    
    def function(self, a, b):
        print('customized func')
        c = a*b
        
        return c
```

```python
testclass = Test()
```

```python
testclass()
```

```python
testclass.__call__()
```

```python
testclass.function(10,16)
```

# 4. applying classmethod decorator 

```python
class Test2:
    def __init__(self):
        self.a2 = 3
        self.b2 = 6
        print('init2')
        print(f'self instance is memorized: a2={self.a2} b2={self.b2}')
    
    def __call__(self):
        print('call2')
        return Test2.function(self.a2, self.b2)
    
    @classmethod
    def function(self, a, b):
        print('customized func2')
        c = a*b
        
        return c
```

# 5. applying static method

```python
class Test5:
    def __init__(self):
        self.a5 = 4
        self.b5 = 5
        print('init5')
        print(f'class instance is memorized: a5={self.a5} b5={self.b5}')
    
    def __call__(self):
        print('call5')
        return Test5.function(self.a5, self.b5)
    
    @staticmethod
    def function(a, b):
        print('customized func5')
        c = a*b
        
        return c
```

```python
t5 = Test5()
```

```python
t5()
```

```python
staticfunc = Test5.function
```

# 7. make more simple and clear of class and static method

```python
class FinalTest:
    def __init__(self, c):
        self.valc = c
        self.vala = 10
        self.valb = 12
        
        print(f'inint and a={self.vala}, b={self.valb}, c={self.valc}')
        
    def __call__(self, d):
        print(f'call and d={d}')
        return FinalTest.function(self.vala, self.valb, self.valc, d)
        
    @staticmethod
    def function(a,b,c,d):
        result = a+b+c+d
        return result
    
    @classmethod
    def function(self, a,b,c,d):
        result = a*b*c*d
        return result
        
```

```python
ft = FinalTest(3)
```

```python
ft(4)
```

```python
staticfunc = FinalTest.function
```

```python
staticfunc(5,6,7,8)
```

```python
class Recorder:
    def __init__(self):
        self._weights = []
        for i in range(0, 2):
            self._weights.append(1)
        print(self._weights[-1])
        print(self._weights[-2])
        print("no. above is from __init__")

    def __call__(self, t):
        self._weights = [self._weights[-1], self._weights[-1] + self._weights[-2]]
        print(self._weights[-1])
        print("no. above is from __call__")
```

```python
weight_recorder = Recorder()
```

```python
for i in range(0, 10):
    weight_recorder(i)
```
