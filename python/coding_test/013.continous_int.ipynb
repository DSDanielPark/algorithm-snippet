#### 1 Problem `1`
- 연속하는 자연수로 n을 만들 수 있는 경우의 수
- https://koreapy.tistory.com/1766

[example]

 


15 - (1) / 1  >>> 나머지 0

15 - (1+2) / 2  >>> 나머지 0

15 - (1+2+3) / 3 >> 나머지 0

15 - (1+2+3+4) / 4 >> 5/4        → 불가능

15 - (1+2+3+4+5) / 5 >> 나머지 0

1.1 My Solution

```python
n=5
```

```python
cnt=0
cnt_sum = 0
answer = 0
for i in range(1,n+1):
    cnt_sum += cnt
    cnt += i
    print(f'({n} - {cnt}) % {i})')
    if cnt > n:
        break
    if ((n - cnt) % i) == n or ((n - cnt) % i) == 0:
        print(i)
        answer += 1
```
