1.1 My Solution


```python
import numpy as np

def get_outline_rotation(matrix, query):
    matrix = np.array(matrix)
    
    #1.상하좌우 특정
    up = matrix[(query[0]-1):query[0], (query[1]-1):query[3]][0]
    up_right = up[-1]
    bot = matrix[(query[2]-1):query[2], (query[1]-1):query[3]][0]
    bot_left = bot[-1]
    left = matrix[:,query[1]-1][query[0]-1:query[2]]
    left_up = left[1]
    right = matrix[:,query[3]-1][query[0]-1:query[2]]
    right_down = right[-2]
    
    
    #2.상하좌우 로테이션수행
    new_up_list = [left_up]
    for i in range(len(up)-1):
        new_up_list.append(up[i])
    bot_list = list(bot)
    bot_list.pop(0)
    new_bot_list = bot_list
    new_bot_list.extend([right_down])
    
    new_left_list = list(left)
    new_left_list.pop(0)
    new_left_list.append(new_bot_list[0])
    
    new_right = list(right)
    new_right.pop(-1)
    new_right_list = [new_up_list[-1]]
    new_right_list.extend(new_right)

    #3.치환
    matrix[(query[0]-1):query[0], (query[1]-1):query[3]][0] = np.array(new_up_list)
    matrix[(query[2]-1):query[2], (query[1]-1):query[3]][0] = np.array(new_bot_list)
    matrix[:,query[1]-1][query[0]-1:query[2]] = np.array(new_left_list)
    matrix[:,query[3]-1][query[0]-1:query[2]] = np.array(new_right_list)
    
    changed = new_up_list+new_bot_list+new_left_list+new_right_list
    min_val = min(changed)
    
    return matrix, min_val

## 최종함수
def solution(rows, columns, queries):
    
    answer = []
    matrix = np.array([i for i in range(1, columns*rows+1)]).reshape(rows, columns)
    
    for q in queries:
        matrix, min_val= get_outline_rotation(matrix, q)
        answer.append(min_val)
        
    answer = [int(x) for x in answer]
    
    return answer
```

```python
# 프로그래머스 박솜님
from collections import deque
def solution(rows, columns, queries):
    arr = [[i+columns*j for i in range(1,columns+1)] for j in range(rows)]
    answer, result = deque(), []
    for i in queries:
        a,b,c,d = i[0]-1,i[1]-1,i[2]-1,i[3]-1
        for x in range(d-b):
            answer.append(arr[a][b+x])
        for y in range(c-a):
            answer.append(arr[a+y][d])
        for z in range(d-b):
            answer.append(arr[c][d-z])
        for k in range(c-a):
            answer.append(arr[c-k][b])
        answer.rotate(1)
        result.append(min(answer))
        for x in range(d-b):
            arr[a][b+x] = answer[0]
            answer.popleft()
        for y in range(c-a):
            arr[a+y][d] = answer[0]
            answer.popleft()
        for z in range(d-b):
            arr[c][d-z] = answer[0]
            answer.popleft()
        for k in range(c-a):
            arr[c-k][b] = answer[0]
            answer.popleft()
    return result
```

```python
# 프로그래머스 seyoung755 , brian6484 , 박관우 , yjincheol 님
def solution(rows, columns, queries):
    answer = []

    board = [[i+(j)*columns for i in range(1,columns+1)] for j in range(rows)]
    # print(board)

    for a,b,c,d in queries:
        stack = []
        r1, c1, r2, c2 = a-1, b-1, c-1, d-1


        for i in range(c1, c2+1):

            stack.append(board[r1][i])
            if len(stack) == 1:
                continue
            else:
                board[r1][i] = stack[-2]


        for j in range(r1+1, r2+1):
            stack.append(board[j][i])
            board[j][i] = stack[-2]

        for k in range(c2-1, c1-1, -1):
            stack.append(board[j][k])
            board[j][k] = stack[-2]

        for l in range(r2-1, r1-1, -1):
            stack.append(board[l][k])
            board[l][k] = stack[-2]

        answer.append(min(stack))


    return answer
```
