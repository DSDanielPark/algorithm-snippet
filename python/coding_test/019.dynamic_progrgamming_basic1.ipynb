#### 1 Problem
##### 1.1 Qestion: 간선에 가중치와 방향성이 없는 임의의 루트 있는 트리가 주어졌을 때, 정점 U를 루트로하는 서브트리에 속한 정점의 수를 출력하라.

##### 1.2 Descriptions:
- 트리의 정점의 수 N
- 루트 번호 R
- 쿼리의 수 Q <br>
(2 ≤ N ≤ 105, 1 ≤ R ≤ N, 1 ≤ Q ≤ 105) 
<br>
- 이어서  N-1줄에 걸쳐, U V의 형태로 트리에 속한 간선의 정보가 주어짐 (1 ≤ U, V ≤ N, U ≠ V) 
- U와 V를 양 끝점으로 하는 간선이 트리에 속함을 의미
- Q줄에 걸쳐, 문제에 설명한 U가 하나씩 주어짐 (1 ≤ U ≤ N) 
- 입력으로 주어지는 트리는 항상 올바른 트리임이 보장됨
<br>

##### 1.3 input <br>
- 아래 query에서 9의 경우 9개의 간선을 의미하고, 그 이후 두개의 숫자는 간선의 양 끝 번호를 의미 (간선을 노드라고 칭하면, 노드간 연결)
- [query] <br>
9 <br>
5 3 <br>
1 3 <br>
4 3 <br>
5 4 <br>
5 6 <br>
6 7 <br>
2 3 <br>
9 6 <br>
6 8 <br>

##### 1.4 output <br>
- input=5     >>     result=9 <br>
- input=4     >>     result=4 <br>
- input=1     >>     result=1 <br>
