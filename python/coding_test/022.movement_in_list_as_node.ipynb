#### 1 Problem
- 리스트의 적힌 값대로 이동해서 -1일경우 종료하게 된다.
- 다음과 같은 A리스트가 존재할 경우, A[0]=1 이므로 A[1]노드로 옮겨가고, A[1]노드의 값은 4이므로, A[4]로 이동, A[4]=2이므로 A[2]=-1이고 드디어 -1이 등장했으므로 이 프로세스를 종료함
  A[0] =  1 <br>
  A[1] =  4 <br>
  A[2] = -1 <br>
  A[3] =  3 <br>
  A[4] =  2 <br>

  - return은 위 상기에서 -1까지 가기까지 총 4번의 값을 확인했으므로, 4를 리턴

#### 2 Solutions


```python
def solution(A):
    cnt=0
    if A[0]==-1:
        return 1
    else:
        move = A[0]
        cnt=1
        for i in range(len(A)):
            if A[move] == -1:
                cnt+=1
                break|
            else:
                move = A[move]
                cnt+=1

    return cnt
```
