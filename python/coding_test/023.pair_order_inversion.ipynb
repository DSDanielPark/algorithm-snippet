An array A consisting of N integers is given. An inversion is a pair of indexes (P, Q) such that P < Q and A[Q] < A[P].

Write a function:

def solution(A)

that computes the number of inversions in A, or returns −1 if it exceeds 1,000,000,000.

For example, in the following array:

 A[0] = -1 A[1] = 6 A[2] = 3
 A[3] =  4 A[4] = 7 A[5] = 4
there are four inversions:

   (1,2)  (1,3)  (1,5)  (4,5)
so the function should return 4.

```python
A = [-1,6,3,4,7,4]
```

```python
#전략1: 뒤에 있는 요소중에 내 요소보다 작은 것들의 수를 카운트함
```

```python
A[0]
A[1:]
```

```python
print(A[4])
print(A[5:])
```

```python
cnt=0
for i in range(len(A)-1):
    cnt += len([x for x in A[i+1:] if x<A[i]])
    
```

```python
def solution(A):
    cnt=0
    for i in range(len(A)-1):
        cnt += len([x for x in A[i+1:] if x<A[i]])

    return cnt
```

```python
solution(A)
```

```python
#엣지케이스 생각
B=[3]
C=[1,2,3,4,5]
```

```python
solution(B)
```
