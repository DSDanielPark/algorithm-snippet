1.1 My Solution


```python
import re

string = "100-200*300-500+20"

int_str = re.sub(r"[^0-9]", " ", string)


int_str = int_str.split(' ')
int_str
```

```python

symbol_str = re.sub(r"[0-9]", "", string)
symbol = symbol_str.split(' ')
```

```python

symbol = list(symbol[0])
symbol
```

```python

new_list = []
for i, sy in enumerate(symbol):
    new_list.append(int_str[i])
    new_list.append(sy)
new_list.append(int_str[-1])

new_list

symbol_list = list(set(symbol))

from itertools import permutations
```

```python

prior = list(permutations(symbol_list, 3))

prior
```

```python

prior[0]
```

```python

symbol_loc = list(filter(lambda x: new_list[x] == '-', range(len(new_list))))
symbol_loc

def cal(a,symbol,b):
    if symbol == '+':
        return int(a)+int(b)
    elif symbol == '*':
        return int(a)*int(b)
    elif symbol == '-':
        return int(a)-int(b)
    
```

```python

def get_cal_list(target_list, prior_symbol):
    symbol_loc = list(filter(lambda x: new_list[x] == prior_symbol, range(len(new_list))))
    
    if len(symbol_loc)==0:
        return None
    else:
        cal_result = cal(target_list[int(symbol_loc[0]-1)], target_list[int(symbol_loc[0])], target_list[int(symbol_loc[0]+1)])
        target_list[symbol_loc[0]-1] = cal_result
        target_list.pop(symbol_loc[0])
        target_list.pop(symbol_loc[0])
    
    return target_list

get_cal_list(new_list, '-')
```
