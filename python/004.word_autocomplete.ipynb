#### 1. Problem Example 
- 자동완성이 완성되기까지 걸리는 횟수 카운트
[example]
- `["I 16", "I -5643", "D -1", "D 1", "D 1", "I 123", "D -1"]` to `[0,0]`
- `["I -45", "I 653", "D 1", "I -642", "I 45", "I 97", "D 1", "D -1", "I 333"]` to `[333, -45]`


```python
def solution(words):
    
    words_len = [len(w) for w in words]
    result = 0
    for i in range(len(words)):
        words_list = words.copy()
        words_list.pop(i)
        
        for j, alphabet in enumerate(words[i]):
            loc_alphabet_list = [word[j:j+1] for word in words_list]
            if alphabet in loc_alphabet_list:
                result += 1
            elif alphabet not in loc_alphabet_list:
                result += 1
                break
                
    return result
출처: https://koreapy.tistory.com/1735 [py:티스토리]
```
